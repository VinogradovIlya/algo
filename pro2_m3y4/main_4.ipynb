```python
import pandas as pd
```

```python
df = pd.read_csv('./DataAnalyst.csv')
```

```python
df['SQL'] = df['Job Description'].apply(lambda x: int('sql' in x.lower()))
df['Python'] = df['Job Description'].apply(lambda x: int('python' in x.lower()))
df['R'] = df['Job Description'].apply(lambda x: int('r' in x.lower()))
df['Microsoft Outlook'] = df['Job Description'].apply(lambda x: int('microsoft outlook' in x.lower()))
df['Excel'] = df['Job Description'].apply(lambda x: int('excel' in x.lower()))
df['Word'] = df['Job Description'].apply(lambda x: int('word' in x.lower()))
df['Mixpanel'] = df['Job Description'].apply(lambda x: int('mixpanel' in x.lower()))
df['MS Power BI'] = df['Job Description'].apply(lambda x: int('ms power bi' in x.lower()) )
df['SharePoint'] = df['Job Description'].apply(lambda x: int('sharepoint' in x.lower()))
df['ServiceNow'] = df['Job Description'].apply(lambda x: int('servicenow' in x.lower()))
df['MS Project'] = df['Job Description'].apply(lambda x: int('ms project' in x.lower()))
df['Planview'] = df['Job Description'].apply(lambda x: int('planview' in x.lower()) )
df['Java'] = df['Job Description'].apply(lambda x: int('java' in x.lower()) )
df['C#'] = df['Job Description'].apply(lambda x: int('c#' in x.lower()) )
df['C++'] = df['Job Description'].apply(lambda x: int('c++' in x.lower()))
```

```python
new_df = df.drop(columns=['Rating', 'Company Name', 'Location', 'Headquarters', 'Size', 'Founded', 'Type of ownership', 'Industry', 'Sector', 'Revenue', 'Competitors', 'Easy Apply'])
new_df
```
