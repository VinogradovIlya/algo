````python
df['Total calories'] = df['Calories'] + df['Calories from Fat']
````

````python
filter_df = df[(df['Total calories'] <= 900)]
filter_df
````

````python
res = pd.merge(filter_df['Item'], filter_df['Item'], how='cross') # !
res
````

````python
res_cal = pd.merge(filter_df['Total calories'], filter_df['Total calories'], how='cross')
res_cal
````

````python
res_with_cal = pd.merge(res, res_cal, left_index=True, right_index=True)
res_with_cal
````

````python
res_with_cal = res_with_cal.assign(Sum=res_with_cal['Total calories_x'] + res_with_cal['Total calories_y'])
res_with_cal
````

````python
res_with_cal.drop(columns=['Total calories_x', 'Total calories_y'], inplace=True)
res_with_cal
````

````python
res2 = res_with_cal[(res_with_cal['Sum'] <= 900) & (res_with_cal['Sum'] >= 600)]
res2
````

### 2 способ

````python
products = dict(zip(filter_df['Item'], filter_df['Total calories']))
products
````

````python
# создаем генератор
"""
https://skillbox.ru/media/code/generatory_python_chto_eto_takoe_i_zachem_oni_nuzhny/

https://habr.com/ru/articles/560300/
"""

def new_product():
    for i in products:
        yield i

np = new_product()
print(next(np))
print(next(np))
print(next(np))
````

````python
# ``` Python
from itertools import combinations
prod = list(products.keys())

# res = []
comb_dict = dict()

for i in range(1, 5):
    res = []
    for comb in combinations(prod, i):
        total = 0
        if i != 1:
            for item in comb:
                total += products[item]
            if 600 <= total <= 900:
                res.append(comb)
        else:
            if 600 <= products[comb[0]] <= 900:
                res.append(comb)
    comb_dict[i] = res

print(len(res))
# код выполняется больше 12 часов
# ```
````
